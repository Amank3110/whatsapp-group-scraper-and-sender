```python
group_invite_link = "https://chat.whatsapp.com/KVB1dxKFGXGFzh44g58ffs"
message = f"""Hi,
Please join this official WhatsApp group using the link below.

The group will be used for important communications such as:
- Internal Job Postings (IJP)
- Training Sessions
- Official Announcements

{group_invite_link}

Thanks & Regards,
"""
```

```python
import random
import urllib
numbers = df['Mobile Number'].to_list()
encoded_message = urllib.parse.quote(message)
for num in numbers:
    driver.get(f"https://web.whatsapp.com/send?phone={num}&text={encoded_message}")
    
    try:
        send_btn = WebDriverWait(driver, 30).until(
            EC.element_to_be_clickable((By.XPATH, "//span[@data-icon='wds-ic-send-filled']"))
        )
        send_btn.click()
        print(f"✅ Message sent to {num}")      
    except Exception as e:
        print(f"❌ Failed to send to {num}: {e}")    
    time.sleep(random.randint(4,5))
print("All messages processed.")
```
